```python
import plotly.graph_objects as go

labels = list(pd.concat([df["mission"], df["rocket"], df["outcome"]]).unique())

label_index = {label:i for i,label in enumerate(labels)}

source = []
target = []
value = []

group1 = df.groupby(["mission","rocket"]).size().reset_index(name="count")

for _,row in group1.iterrows():
    source.append(label_index[row["mission"]])
    target.append(label_index[row["rocket"]])
    value.append(row["count"])

group2 = df.groupby(["rocket","outcome"]).size().reset_index(name="count")

for _,row in group2.iterrows():
    source.append(label_index[row["rocket"]])
    target.append(label_index[row["outcome"]])
    value.append(row["count"])
```

```python
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        label=labels
    ),
    link=dict(
        source=source,
        target=target,
        value=value
    )
)])
fig.update_layout(title="Soviet Space Dogs Missions Sankey Diagram")
fig.show()
```
